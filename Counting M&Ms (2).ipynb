```
import requests, io
```

```
import pandas
```

```
git_url = "https://raw.githubusercontent.com/databricks/LearningSparkV2/master/databricks-datasets/learning-spark-v2/mnm_dataset.csv"
```

```
raw_data = requests.get(git_url).content
```

```
mnm_pd = pandas.read_csv(io.StringIO(raw_data.decode("UTF-8")))
```

##### Find the count of MnM candies for each color and state and order them in descending order

```
mnm_df = spark.createDataFrame(mnm_pd)
```

```
from pyspark.sql.functions import col, count
```

```
candy_count = mnm_df.groupBy('State', 'Color') \
                    .agg(count('Count').alias('Total_Candy_Count')) \
                    .orderBy('Total_Candy_Count', ascending = False)
```
