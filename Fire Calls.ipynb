####Working on Fire Incidents dataset

```
import requests, io, pandas
```

```
git_url = "https://raw.githubusercontent.com/databricks/LearningSparkV2/master/chapter3/data/sf-fire-calls.csv"
```

```
fire_pd = pandas.read_csv(git_url)
```

```
fire_pd
```

```
from pyspark.sql.types import *
```

```
fire_schema = StructType([StructField('CallNumber', IntegerType(), True),
 StructField('UnitID', StringType(), True),
 StructField('IncidentNumber', IntegerType(), True),
 StructField('CallType', StringType(), True),
 StructField('CallDate', StringType(), True),
 StructField('WatchDate', StringType(), True),
 StructField('CallFinalDisposition', StringType(), True),
 StructField('AvailableDtTm', StringType(), True),
 StructField('Address', StringType(), True),
 StructField('City', StringType(), True),
 StructField('Zipcode', FloatType(), True),
 StructField('Battalion', StringType(), True),
 StructField('StationArea', StringType(), True),
 StructField('Box', StringType(), True),
 StructField('OriginalPriority', StringType(), True),
 StructField('Priority', StringType(), True),
 StructField('FinalPriority', IntegerType(), True),
 StructField('ALSUnit', BooleanType(), True),
 StructField('CallTypeGroup', StringType(), True),
 StructField('NumAlarms', IntegerType(), True),
 StructField('UnitType', StringType(), True),
 StructField('UnitSequenceInCallDispatch', FloatType(), True),
 StructField('FirePreventionDistrict', StringType(), True),
 StructField('SupervisorDistrict', StringType(), True),
 StructField('Neighborhood', StringType(), True),
 StructField('Location', StringType(), True),
 StructField('RowID', StringType(), True),
 StructField('Delay', FloatType(), True)])
```

```
fire_df = spark.createDataFrame(fire_pd, schema = fire_schema)
```

```
display(fire_df)
```

#####1. How many distinct CallTypes were recorded as the causes of the fire calls?

```
from pyspark.sql.functions import countDistinct, col, to_timestamp, count
```

```
fire_callTypes = fire_df.select("CallType") \
                        .agg(countDistinct("CallType").alias("DistinctCallTypes"))
display(fire_callTypes)
```

##### 2. Change the name of Delay column to ResponseDelayedinMins and take a look at the response times that were longer than five minutes.

```
fire_response_delay = fire_df.withColumnRenamed('Delay', 'ResponseDelayedinMins') \
                             .where(col('ResponseDelayedinMins') > 5.0)
display(fire_response_delay)
```

##### 3. Convert columns CallDate, WatchDate, and AlarmDtTm to TimeStamp type

```
fire_datecols = fire_df.withColumn("IncidentDate", to_timestamp(col("CallDate"), "MM/dd/yyyy")) \
                       .drop("CallDate") \
                       .withColumn("OnWatchDate", to_timestamp(col("WatchDate"), "MM/dd/yyyy")) \
                       .drop("WatchDate") \
                       .withColumn("AvailableDtTS", to_timestamp(col("AvailableDtTm"), "MM/dd/yyyy hh:mm:ss a")) \
                       .drop("AvailableDtTm")
```

##### 4. What are the most common types of fire calls?

```
fire_common_calls = fire_df.select(col('CallType')) \
                           .groupBy(col('CallType')) \
                           .agg(count(col('CallType')).alias('CallType_Count')) \
                           .orderBy(col('CallType_Count'), ascending = False)
```

```
display(fire_common_calls)
```

#####5. What zip codes accounted for the most calls?

```
most_zip_calls = fire_df.select(col('Zipcode')) \
                        .groupBy(col('Zipcode')) \
                        .agg(count(col('Zipcode')).alias('Zipcode_CallCounts')) \
                        .orderBy(col('Zipcode_CallCounts'), ascending = False)
```

```
display(most_zip_calls)
```

##### 6. Compute the sum of alarms, the average response time, and the minimum and maximum response times to all fire calls in our data set

```
from pyspark.sql.functions import avg, min, max
```

```
fire_stats = fire_df.select(sum(col('Delay')), avg(col('Delay')), min(col('Delay')), max(col('Delay')))
```

```
display(fire_stats)
```

##### 8. What months within the year 2018 saw the highest number of fire calls?

```
from pyspark.sql.functions import *
```

```
fire_max_2018 = fire_calls_date.where(year(col('CallDate')) == 2018) \
                               .groupBy(month(col('CallDate')).alias('Month')) \
                               .agg(count(col('CallNumber')).alias('FireCallsCount')) \
                               .orderBy(col('FireCallsCount'), ascending = False)
```

```
display(fire_max_2018)
```

##### 9. Which neighborhood in San Francisco generated the most fire calls in 2018?

```
fire_neighborhood_2018 = fire_calls_date.where(year(col('CallDate')) == 2018) \
                                        .groupBy(col('neighborhood')) \
                                        .agg(count(col('CallNumber')).alias('FireCallsCount')) \
                                        .orderBy(col('FireCallsCount'), ascending = False)
```

```
display(fire_neighborhood_2018)
```
